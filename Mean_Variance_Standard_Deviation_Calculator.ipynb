Screenshot 2024-11-17 211840.pngDictionary returned should follow this format:

{
  'mean': [axis1, axis2, flattened],
  'variance': [axis1, axis2, flattened],
  'standard deviation': [axis1, axis2, flattened],
  'max': [axis1, axis2, flattened],
  'min': [axis1, axis2, flattened],
  'sum': [axis1, axis2, flattened]
}

```python
import numpy as np
import pandas as pd
```

```python
def calculate(list):
    if len(list) != 9:
        raise ValueError("List must contain nine numbers.")
    df = np.reshape(list, (3,3))
    calculations = {'mean':[df.mean(axis = 0).tolist(), df.mean(axis =1).tolist(), df.mean().tolist()],
                 'variance':[df.var(axis = 0).tolist(), df.var(axis =1).tolist(), df.var().tolist()],
                 'standard deviation':[df.std(axis = 0).tolist(), df.std(axis =1).tolist(), df.std().tolist()],
                 'max':[df.max(axis = 0).tolist(), df.max(axis =1).tolist(), df.max().tolist()],
                 'min':[df.min(axis = 0).tolist(), df.min(axis =1).tolist(), df.min().tolist()],
                 'sum':[df.sum(axis = 0).tolist(), df.sum(axis =1).tolist(), df.sum().tolist()]}
    return calculations
```

```python
list = (1,2,3,4,5,6,7,8,9)
calculate(list)
```
